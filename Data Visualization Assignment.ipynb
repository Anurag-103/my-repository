## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),linestyle='dashdot',color='black')
plt.xlabel('X',size=15)
plt.ylabel("Y",size=15)
plt.title('Sine Wave Graph',size=15)
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
S=['Delhi',"Mumbai",'Channai',"Kolkata",'Hyderabad']
clothing=np.array([81,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
var=[x for x,_ in enumerate(S)]
plt.bar(var,clothing,width=0.8,label="clothing",color='red',bottom=Equipment+Accessories)
plt.bar(var,Equipment,width=0.8,label="Equipment",color='yellow',bottom=Accessories)
plt.bar(var,Accessories,width=0.8,label="Accessories",color='green')
plt.ylabel('Items',size=15)
plt.xlabel("States",size=15)
plt.xticks(var,S,rotation=90)
plt.title("RETAILER DATABASE",size=15)
plt.grid(False)
plt.legend()
plt.show()
```

## Q.5 For Titanic Dataset

- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
t=sns.load_dataset('titanic')
```

```python
t.head()
sns.barplot(x='class',y='survived',hue='class',palette='Accent',data=t)
plt.legend(loc='upper right')
plt.show()
```

## Q.6 For Tips Dataset
